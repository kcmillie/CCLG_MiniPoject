```python
import pandas as pd
import numpy as np
import csv
csvface = 'Facebook.csv'
csvgoog = 'Google.csv'
csvyelp = 'Yelp.csv'
```

```python
facebook_pd = pd.read_csv(csvface)
google_pd = pd.read_csv(csvgoog)
yelp_pd = pd.read_csv(csvyelp)
merge_table1=pd.merge(facebook_pd, yelp_pd, on='Name')
merge_table1
```

```python
avg_rating = merge_table1.mean(axis=1)
avg_rating
```

```python
merge_table1['Average Rating'] = avg_rating
del merge_table1["avg_rating"]
merge_table1
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
sns.swarmplot(x = 'Average Rating', y = 'Name', data = merge_table1)
plt.savefig("Average Ratings.PNG", bbox_inches='tight')
```
